### 가설) 최근 학습 성취도가 높을수록 구독 연장 비율이 높을 것이다.

````python
# 필요한 데이터 칼럼 생성
df6 = access_data[ ['recent_learning_achievement', 'user_id', 'target'] ]
df6
````

````python
# 성취도를 구간별로 구분하기
# 범주화할 데이터 리스트 만들기
achive_rate = list(df6['recent_learning_achievement'])
achive_rate[:10]
````

````python
# 범주를 구분하는 기준 및 라벨 만들기
bins = [30, 60, 70, 80, 90, 120]
labels=['E', 'D', 'C', 'B', 'A']

# 성취도에 따라 등급별로 변경하기
cuts = pd.cut(achive_rate, bins, right = True, include_lowest=True, labels=labels)
cuts
````

````python
# 범주화 결과물을 테이블로 변경
cuts = pd.DataFrame(cuts)
cuts.tail()
````

````python
# 기존 테이블에 합치기
df6 = pd.concat( [df6, cuts], axis = 1, join='inner')
df6.head()
````

````python
# 컬럼명 변경
df6.columns = ['recent_learning_achievement', 'user_id', 'target', 'levels']
df6
````

````python
# 최근 성취도별 구독 연장 평균 비율
subscription_extension_rate6 = df6.groupby('levels')['target'].mean()
subscription_extension_rate6
````

````python
# 시각화
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib

plt.figure(figsize=(8,6))
plt.bar(subscription_extension_rate6.index, subscription_extension_rate6.values)
plt.title('최근 학습 성취도에 따른 다음달 구독 연장 비율')
plt.xlabel('학습 성취도')
plt.ylabel('구독 연장 여부(%)')
plt.ylim(0.5,0.7)
plt.yticks([0.5, 0.6, 0.7])
#plt.xticks(subscription_extension_rate6, ['A', 'B', 'C', 'D', 'E'])

plt.show()
````

* 학습 성취도가 높을수록 구독 비율 높음
* 학습 성취도가 가장 낮은 그룹 역시 구독 비율이 높음
