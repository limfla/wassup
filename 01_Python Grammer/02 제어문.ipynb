# while 문

* 조건에 해당하면(조건문이 참이면) 무한 반복
* 조건에 해당하는 동안 다음 동작을 계속 실행해줘

```python
treehit = 0
while treehit < 10:
    treehit = treehit +1
    print("나무를 %d번 찍었습니다." % treehit)
    if treehit == 10:
        print("나무 넘어갑니다.")
```

## continue - while문의 맨 처음으로 돌아가기

```python
# 1부터 10까지의 숫자 중에서 홀수만 출력

a = 0 # 초기값을 지정해줘야 조건문 비교 가능

while a < 10:
    a = a + 1
    if a % 2 == 0 : continue # 짝수이면 처음으로 돌아가
    print(a)

# % - 나머지 , // - 몫
```

```python
# a가 1부터 시작한다면

a = 1
while a <= 10:
    if a % 2 == 0: a = a + 1; continue
    print(a)
    a = a + 1
    
# 여러 줄을 한 줄로 쓸 때는 ;(콜론)
```

## 전형적인 for문

* 시퀀셜객체-리스트, 튜플, 문자열(순서가 있는)의 값을 차례로 뽑아서 아래 문장을 반복 수행해줘

```python
test_list = ['one', 'two', 'three']
for i in test_list:
    print(i)

# 리스트에서 하나씩 뽑아서 변수로 출력
```

## for문의 응용

```python
# 총 5명의 학생이 시험을 보았는데 시험 점수가 60점 이상이면 합격이고 그렇지 않으면 불합격이다.
# 합격인지, 불합격인지 결과를 보여 주시오.

marks = [90, 25, 67, 45, 80] # 학생들의 시험 점수 리스트
number = 0 # 학생에게 붙여 줄 번호

for mark in marks: # 90, 25, 67, 45, 80을 순서대로 mark에 대입
    number = number + 1
    if mark >= 60:
        print("%d번 학생은 합격" % number)
    else:
        print("%d번 학생은 불합격" % number)
```

## for문과 continue문

```python
# 60점 이상인 사람에게는 축하 메시지를 보내고 나머지 사람에게는 아무런 메시지도 전하지 않는

marks = [90, 25, 67, 45, 80] # 학생들의 시험 점수 리스트
number = 0 # 학생에게 붙여 줄 번호

for mark in marks:
    number = number + 1
    if mark < 60:
        continue
    print("%d번 학생 축하합니다. 합격입니다." % number)
```

## 리스트 컴프리헨션

* 리스트 안에 for문을 포함

```python
a = [1,2,3,4]
result = [] # 빈 리스트 객체

for num in a :
    result.append(num*3)
print(result)
```

```python
# 리스트 컴프리헨션 사용

a = [1,2,3,4]
result = [num * 3 for num in a] # 리스트 객체에 직접 for문을 넣음 (표현식 + for문)
print(result)
```

```python
# 짝수에만 3을 곱해 담기 (if문도 추가할 수 있다)

a = [1,2,3,4]
result = [num * 3 for num in a if num % 2 == 0]
print(result)
```
