```python
# 제목과 url 데이터 목록 만들기 - 반복문
news_title = []
news_url = []

for i in srclist:
    # 태그 객체의 메서드(.txt)를 쓰려면 태그 객체별로 빼줘야함
    title = i.text
    url = urljoin(PATH, i['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)
```

### 파라미터 처리하기
* get과 post가 어떻게 동작하는지 알아보자

```python
import requests
```

```python
param = {'name':'Amy', 'age':20, 'address':'Seould'}
```

```python
# 요청 정보(파라미터)가 인자에 포함되어 있다.
resp1 = requests.get('http://httpbin.org/get', params=param)
print(resp1) # 오류
print(resp1.text)
```

```python
# 요청 정보(파라미터)가 인자에 없고, form에 숨겨져있다.
resp2 = requests.post('http://httpbin.org/post', data=param)
print(resp2.text)
```
